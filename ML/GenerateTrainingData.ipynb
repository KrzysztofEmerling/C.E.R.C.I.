```python
pgn_path = "../tests/Eval/LichessDB/lichess_db_standard_rated_2016-06.pgn"
output_path_agresive = "./trainingAgresive.txt"
output_path_positional = "./trainingPositional.txt"
def count_games(pgn_path):
    count = 0
    with open(pgn_path, "r", encoding="utf-8") as f:
        for line in f:
            if line.startswith("[Termination"):
                count += 1
    return count

num_games = count_games(pgn_path)
print(f"Liczba partii: {num_games}")
```
